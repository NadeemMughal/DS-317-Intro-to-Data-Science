# Pandas Functions to Modify the Dataset

### copy() Function is used to make a copy of our dataset original dataset is still within df.

```python
fd=df.copy()
```

```python
fd
```

### rename() Function used to rename specific column with specific name AT A TIME multiple column change name here.

```python
df.rename(columns={"Quantity": "NUMBERS","Status":"TARGET"})
```

### where() Function used to find a specific values from dataset other values it will show NaN by Default.

```python
df.where(df["Quantity"]>70)
```

### But  if we replace NaN values with specific value we need add other as an argument.

```python
df.where(df["Quantity"]>70,other=1000000000000)
```

```python
df
```

### drop() Function used to drop any column or row from dataset.
##### axis=1 means column selected.
##### axis=0 means row selected

```python
df.drop("Quantity",axis=1)
```

```python
df.drop(1,axis=0)
```
